## Overfitting and Underfitting 

A major challenge in machine learning is ensuring that a model generalizes well to unseen data. 

### Two common issues are: 
#### ✓ Overfitting: The model learns noise instead of the underlying pattern, performing well on training data but poorly on new data. 
#### ✓ Underfitting: The model is too simple to capture patterns in the data, leading to poor performance even on the training set. 

### Example: Overfitting vs. Underfitting 

Consider a dataset of student exam scores based on study hours. A model that is too simple (e.g., a 
straight line) underfits the data, while a highly complex polynomial model that fits every data point 
overfits.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Generate sample data
np.random.seed(42)
x = np.linspace(0, 10, 100).reshape(-1, 1)
y = np.sin(x).ravel() + np.random.normal(scale=0.5, size=x.shape[0])
```

```python
# Fit a linear model (underfitting)
lin_reg = LinearRegression()
lin_reg.fit(x, y)
y_pred_linear = lin_reg.predict(x)
```

```python
# Fit a polynomial model ()
poly_features = PolynomialFeatures(degree=10)
x_poly = poly_features.fit_transform(x)
lin_reg.fit(x_poly, y)
y_pred_poly = lin_reg.predict(x_poly)
```

```python
# plot results
plt.scatter(x, y, color='blue', label="Data")
plt.plot(x, y_pred_linear, color='green', label="Underfitting (Linear Model)")
plt.plot(x, y_pred_poly, color='red', label='Overfitting (polynomial Model)')
plt.legend()
plt.title("Overfitting vs. Underfitting")
plt.show()
```
