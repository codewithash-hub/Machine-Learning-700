Using the California Housing dataset, demonstrate the steps to train a Linear Regression model including label encoding if necessary.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Load data
data = pd.read_csv("CaliforniaHousing.csv")
data.head()
```

```python
# Seperate features and targets
X = data.drop(columns=["median_house_value", "ocean_proximity"])
y = data["median_house_value"]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Train model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_train)
```

```python
print(f"{y_pred}")
import matplotlib.pyplot as plt

plt.boxplot(y_pred)
```
