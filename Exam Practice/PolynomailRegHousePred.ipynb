```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
X = np.array([1200, 1500, 1800, 2100, 2500]).reshape(-1, 1)
y = np.array([220, 270, 320, 370, 420])

# Train model and make prediction
poly_model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
poly_model.fit(X, y)

# Predict new price for 2000sqft house
new_house = np.array([[2000]])
pred_price_poly = poly_model.predict(new_house)
print(f'Predicted price (Polynomial Regression): ${pred_price_poly[0]:.2f}K')
```

```python
# Visualize Polynomail regression
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, poly_model.predict(X), c='green', label='Polynimial curve')
plt.title("Polynomial Regression: House Price Prediction")
plt.xlabel('Square footage')
plt.ylabel('Price ($1000)')
plt.legend()
plt.show()
```
