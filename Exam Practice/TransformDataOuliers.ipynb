```python
from sklearn.preprocessing import RobustScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Sample skewed data with outliers
df = pd.DataFrame({'value': [1, 2, 3, 4, 5, 100, 120]})

# Log transformation
df['log_value'] = np.log1p(df['value'])

# Robust scaling
scaler = RobustScaler()
df['robust_scaled'] = scaler.fit_transform(df[['value']])

print("Transformed Data:\n", df)

# Optional: Plot the distributions
sns.histplot(df['value'], color='red', label='Original', kde=True)
sns.histplot(df['log_value'], color='green', label='Log Transformed', kde=True)
sns.histplot(df['robust_scaled'], color='blue', label='Robust Scaled', kde=True)
plt.legend()
plt.title('Distribution after Transformations')
plt.show()
```
