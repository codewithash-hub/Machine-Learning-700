Example: Predicting House Prices Using Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
X = np.array([1200, 1500, 1800, 2100, 2500]).reshape(-1, 1)
y = np.array([220, 270, 320, 370, 420])

# Model training and predicting
model = LinearRegression()
model.fit(X, y)
# predicting house price for 2000 sqft house
new_house = np.array([[2000]])
predicted_price = model.predict(new_house)
print(f"Predicted price 2000sqft house: ${predicted_price[0]:.2f}K")
```

```python
# Visualizing the regression line
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.title('Regression Line: Predicting House Price')
plt.xlabel('Square footage')
plt.ylabel('Prices ($1000)')
plt.legend()
plt.show()
```
